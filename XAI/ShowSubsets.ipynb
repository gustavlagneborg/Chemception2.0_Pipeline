```python
import rdkit
from rdkit import Chem
from rdkit.Chem.Draw import rdMolDraw2D
from rdkit.Chem.Draw import IPythonConsole
import sys, os

sys.path.append(os.path.join(os.path.dirname(os.path.abspath(__file__)), os.pardir))


def mol_with_atom_index( mol ):
    atoms = mol.GetNumAtoms()
    for idx in range( atoms ):
        mol.GetAtomWithIdx( idx ).SetProp( 'molAtomMapNumber', str( mol.GetAtomWithIdx( idx ).GetIdx() ) )
    return mol



smi = "O=S(=O)(O)c1cc2ccc1ccc1ccc(cc1S(=O)(=O)O)nnc1ccc(ccc3ccc(cc3S(=O)(=O)O)nn2)c(S(=O)(=O)O)c1"


mol = Chem.MolFromSmiles(smi)
highlight = [
    Chem.MolFromSmarts("*=S(=O)(O)"),
    Chem.MolFromSmarts("*(=O)(-O)")
]



overlap = [mol.GetSubstructMatches(hl)for hl in highlight]
print(overlap)

mol.__sssAtoms = [1,2,3,20,21,37,39]


mol


```
